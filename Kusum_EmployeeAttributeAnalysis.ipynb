```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importing dataset
data = pd.read_csv('employee_data.csv', sep='\t')
```

```python
data.head()
```

```python
data.columns
```

**[1]**

**Continous columns**


```python
num_cols = data._get_numeric_data().columns
cont_cols = []
for i in num_cols:
  if(len(data[i].unique())) > int(data.shape[0]/25):
     cont_cols.append(i)
print (cont_cols)
```

**Categorical columns**

```python
cat_cols = list(set(data.columns) - set(cont_cols))
cat_cols
```

**[2] UNIVARIATE ANALYSIS**

**CATEGORICAL FEATURES**

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(x="education", data=data)
plt.title("Education", fontdict={'fontsize' : 20})
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

*Inference*

Most number of employees are HS-graduates, followed by Some-college and Bachelors.
Least number of employees are Preschool.

```python
plt.figure(figsize=(20,10))
sns.countplot(data.employment_type)
plt.title("Employment Type", fontdict={'fontsize' : 20})
plt.show()
data.employment_type.value_counts()
```

*Inference*

More than 650 employees belong to the Private Sector. Least number of employees blong to Federal Government Sector.

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(x="marital-status", data=data)
plt.title("Marital Status", fontdict={'fontsize' : 20})
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
data['marital-status'].value_counts()
```

*Inference*

Most employees belong to Married-civ-spouse category. Only 1 employee belongs to Marries-AF-spouse category.

```python
sns.countplot(data.Gender)
plt.title("Gender Wise Distribution", fontdict={'fontsize' : 20})
data.Gender.value_counts()
```

*Inference*

The number of female employees are less than half the number of male employees.

```python
plt.figure(figsize=(20,10))
sns.countplot(data.relationship)
plt.title("Relationship", fontdict={'fontsize' : 20})
plt.show()
data.relationship.value_counts()
```

**CONTINUOUS FEATURES**

```python
plt.figure(figsize=(12,5))
plt.hist(data['age'], color='orange')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("AGE", fontdict={'fontsize' : 20})
plt.show()
```

*Inference*

Most employees belong to the age group 30-45. Very few employees belong to the age group of 60-80.

```python
plt.figure(figsize=(12,5))
plt.hist(data['salary'], color='orange')
plt.xlabel("Salary")
plt.ylabel("Count")
plt.title("SALARY", fontdict={'fontsize' : 20})
```

*Inference*

Salaries of most employees range between 1,00,000- 2,00,000.

```python
plt.figure(figsize=(12,5))
plt.hist(data['hours_per_week'], color='orange')
plt.xlabel("Hours per week")
plt.ylabel("Count")
plt.title("HOURS PER WEEK", fontdict={'fontsize' : 20})
```

*Inference*

More than 500 employees work 30-40 hours per week.
